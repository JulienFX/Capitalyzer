## Fichier Token.json : 
Les clés d'authentification contenues dans ce fichier doivent être renouvellées de manière hebdomadaire. Veillez à supprimer le fichier chaque semaine (script gère renouvellement) pour que l'authentification puisse se faire correctement. 

## Dataset Linkedin : 
1er constat : chaque mail de candidature est envoyé par l'adresse mail : jobs-noreply@linkedin.com <br>
2e constat : chaque mail de confirmation de réception de candidature commence par "Prénom, votre candidature a été envoyée" <br>
3e constat : chaque mail de refus vis-à-vis d'une candidature commence par "Votre candidature :" <br>
4e constat : automatisation massive des processus -> on peut très aisément s'en sortir sans jamais utiliser de LLM dans ce cas et uniquement procéder avec des expressions régulières
